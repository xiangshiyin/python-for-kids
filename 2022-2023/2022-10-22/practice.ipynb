## Challenge 6: print all integers that are less than 10 and greater than or equal to 0
 * Follow-up: print all integers that are less than 10, but only the even numbers.

````python
counter = 0
while counter < 10:
    print(counter)
    counter = counter + 1
````

````python
counter = 0
while counter < 10:
    if counter % 2 == 0: # this is an even number
        print(counter)
    counter = counter + 1
````

## Challenge 7: print the first 10 [Fibonacci numbers](https://en.wikipedia.org/wiki/Fibonacci_number)

* we know F0 = 0, F1 = 1, F(n) = F(n-1) + F(n-2)

````python
previous = 0
pprevious = 1
counter = 3

while counter <= 10:
    current = previous + pprevious
    print(current) # gives us the new number
    pprevious = previous
    previous = current
    # update the counter
    counter = counter + 1
````

````python
pprevious = 0
previous = 1
counter = 0

while counter < 5:
    print("before the iteration", previous, pprevious)
    
    if counter == 0:
        print(0)
    elif counter == 1:
        print(1)
    else:
        current = previous + pprevious
        print(current) # gives us the new number
        pprevious = previous
        previous = current
    # update the counter
    counter = counter + 1
````
