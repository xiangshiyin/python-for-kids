```python
pool = list(range(1,101))
pool[:10]
```

```python
# pool = list(range(1,101))
target = 99
l = 0
r = len(pool) - 1
counter = 0

while l <= r:
    m = (l + r) // 2
    counter += 1
    if pool[m] < target:
        l = m + 1
    elif pool[m] > target:
        r = m - 1
    else:
        print('I found it')
        print(f'I found the number after {counter} steps')
        break
```

```python
l, r, m, pool[m]
```

```python
import matplotlib.pyplot as plt
import math

ns = list(range(1,101))
steps_linear = list(range(1,101))
steps_binary = [math.log(n, 2) for n in ns]
plt.plot(ns, steps_linear, label='linear search')
plt.plot(ns, steps_binary, label='binary search')
plt.title("steps vs. length")
plt.xlabel('n (length of the list)')
plt.ylabel('steps (to search the target)')
plt.show()
```
